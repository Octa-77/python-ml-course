```python
import pandas as pd
import urllib3 as urlib
import csv

mainpath = "c:/Users/Arduino-Guy/Desktop/ML_course_2020/python-ml-2020/python-ml-course/datasets/"
filename = "titanic/titanic3.xls"
#medals_url = "http://winterolympicsmedals.com"
medals_url = "http://winterolympicsmedals.com/medals.csv"

http = urlib.PoolManager()
r = http.request('GET',medals_url)
print("El estado de la respuesta es %d" %(r.status))
#response = r.data ## CORREGIDO: eliminado un doble decode que daba error
#response
```

```python
# Decode string to UTF8 and SPLIT by line into a LIST
response = r.data.decode(encoding='UTF-8',errors='strict').split("\n")
response
```

```python
#Uses FIRST LINE on response LIST to get the number of elements (columns)
cols = response[0].strip().split(",")
nr_cols = len(cols)

print(response[0])
print (cols[0],nr_cols)
```

```python

row_counter = 0

main_dict = {}
for col in cols:
    main_dict[col] = []
for line in response:
    if row_counter == 0 : 
            row_counter +=1
            continue
    values = line.strip().split(",")
    for i in range(nr_cols):
        main_dict[cols[i]].append(values[i])
    row_counter += 1
    
print("El data set tiene %d filas y %d columnas and it looks like this : \n"%(row_counter-1, nr_cols))
```
